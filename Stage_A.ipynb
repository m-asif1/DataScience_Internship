#### Analysis of a given Data Set (Food Balance Sheet)

```python
# Code to read csv file
df = pd.read_csv("FoodBalanceSheets_E_Africa_NOFLAG.csv",encoding='latin-1')
df.head()
```

```python
# Total number of unique countries in the dataset

df.groupby('Area').nunique().sum()['Area Code']
```

```python
# least correlation with ‘Element Code’
df.corr()["Element Code"]["Y2014":]
```

```python
#  Finding Element that had highest sum in 2017

df[['Area','Y2017']].groupby('Area').sum().sort_values(by='Y2017', ascending = False)
```

```python
# Finding the Element that had the 7th lowest sum in 2017

df.groupby("Area")["Y2017", "Area"].sum().sort_values(by = "Y2017", ascending = True)
```

```python
# Finding the total Protein supply quantity in Madagascar in 2015

df.groupby(['Element', 'Area']).sum().loc[['Protein supply quantity (g/capita/day)']]['Y2015'].reset_index()[['Area','Y2015']].iloc[[23]]
```

```python
# Question 13 
# Finding the percentage of missing data in 2014 to 3 decimal places

missing_data = df["Y2014"] .isna().sum()
percentage = (missing_data / df.shape[0]) * 100
missing_data, round(percentage, 3)
```

```python
# Getting the highest sum of Stock Variation
years= ['Y2014', 'Y2015', 'Y2016','Y2017', 'Y2018']
df.groupby('Element').sum().loc[['Stock Variation'],years].idxmax(axis=1).reset_index(drop=True)
```

```python
# Finding total number of the sum of Processing in 2017

df.groupby("Element").sum().loc["Processing", "Y2017":]
```

```python
# Finding  the mean and standard deviation across the whole dataset
# for the year 2017 to 2 decimal places

round(df.describe().loc[['mean','std'], ['Y2017']], 2)
```

```python
# Finding the total sum of Wine produced in 2015 and 2018

df.groupby('Item').sum().loc[['Wine'],['Y2015','Y2018']]
```
