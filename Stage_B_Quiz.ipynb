```python
# Import the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read the data
energy_df= pd.read_csv('energydata_complete.csv')
energy_df.head()
```

```python
energy_df.shape
```

```python
energy_df.describe()
```

```python
# checking the missing data
energy_df.isnull().sum()
```

```python
#drop the 'date' and 'lights' columns
energy_df = energy_df.drop(columns=['date', 'lights'], axis=1)
energy_df
```

```python
#Normalise the dataset 
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

normalised_energy_df = pd.DataFrame(scaler.fit_transform(energy_df), columns = energy_df.columns)
normalised_energy_df
```

```python
# spliting the Dataset

X = normalised_energy_df.drop(['Appliances'], axis = 1)
y = normalised_energy_df['Appliances']
```

```python
# training & testing spliting the  data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)

print('X_train shape: {}'.format(X_train.shape))
print('X_test shape: {}'.format(X_test.shape))
print('y_train shape: {}'.format(y_train.shape))
print('y_test shape: {}'.format(y_test.shape))
```

```python
# choose sample from dataset
reg_energy_df = normalised_energy_df[['T2', 'T6']]
reg_energy_df.head()
```

```python
# Reshape the sample of dataset
xa = reg_energy_df['T2'].values.reshape(-1,1)
ya = reg_energy_df['T6'].values.reshape(-1,1)
```

```python
# training & testing spliting sample of data
xa_train, xa_test, ya_train, ya_test = train_test_split(xa, ya, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
linear_model = LinearRegression()

# fiting the model to training dataset
linear_model.fit(xa_train, ya_train)

# for predictions
predicted_values = linear_model.predict(xa_test)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(X_train, y_train)
y_pred = reg.predict(X_test)
train_score = reg.score(X_train, y_train)
test_score = reg.score(X_test, y_test)
print('Training score: {:.3f}'.format(train_score))
print('Testing score: {:.3f}'.format(test_score))
```

**What is the R^2 value in two d.p**

```python

from sklearn.metrics import r2_score
r2_scor = r2_score(ya_test, predicted_values)
print('R-squared value in two d.p.:', round(r2_scor, 2))
```

**Mean Absolute Error (MAE)**

```python
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test, y_pred)
print('Mean Absolute Error: {}'.format(round(mae, 2)))
```

**Residual Sum of Squares(RSS)**

```python
rss = np.sum(np.square(y_test - y_pred))
print('Residual Sum of Squares(RSS):', round(rss, 2))
```

```python
# Root Mean Squared Error (RMSE)
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
print('Root Mean Squared Error: {}'.format(round(rmse, 3)))
```

**coefficient of determination**


```python
#from sklearn.metrics import r2_score
r_squared = r2_score(y_test, y_pred)
print('Coefficient of Determination : {}'.format(round(r_squared, 2)))
```

**Which features have the lowest and highest weights respectively**

```python
def get_weights_df(model, feat, col_name):
    
  #this function returns the weight of every feature
    weights = pd.Series(model.coef_, feat.columns).sort_values()
    weights_df = pd.DataFrame(weights).reset_index()
    weights_df.columns = ['Features', col_name]
    weights_df[col_name].round(3)
    
    return weights_df
```

**Which features have the lowest and highest weights respectively**

```python
linear_reg_weights = get_weights_df(reg, X_train, 'Linear_model_Weight')
linear_reg_weights
```

RH_2 feature has the lowest weight, and RH_1 feature has the highest weight

**Ridge regression model alpha value 0.4.**            
**Is there any change to the root mean squared error (RMSE)**
